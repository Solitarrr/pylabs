```python
import random
# генерация массива
A = []
n = int(input("Введите кол-во элементов ", ) or "20")
min_limit = int(input("Введите минимальное значение генерируемых чисел: ", ) or "-20")
max_limit = int(input("Введите максимальное значение генерируемых чисел: ", ) or "20")
if min_limit >= max_limit:
    print("Минимальное не может быть больше максимального. Были взяты значения по умолчанию!")
    min_limit = -20
    max_limit = 20
for i in range(n):
    A.append(random.randint(min_limit,max_limit))
A.sort()
print("\nСгенерирован начальный набор случайных данных: ", A)

# линейный поиск
x = int(input("Введите элемент, который хотите найти с помощью функции линейного поиска: ", ))
def linear_search(arr, x):
    for i in range(len(arr)):
        if arr[i] == x:
            return i
    return 'В сгенерированном массиве нет элемента с таким значением'

print(linear_search(A,x))

x = int(input("Введите элемент, который хотите вставить: ", ))
# функция вставки элемента
def insert_insert(arr, x):
    arr.append(x)
    arr.sort()
    return arr

print("Вывод сортированного массива со вставленным элементом: ")
print(insert_insert(A, x))

x = int(input("Введите элемент, который хотите удалить: ", ))
# функция удаления элемента
def delete_delete(arr, x):
    arr.remove(x)
    arr.sort()
    return arr

print("Вывод сортированного массива с удаленным элементом: ")
print(delete_delete(A, x))

# бинарный поиск
y = int(input("Введите элемент, который хотите найти с помощью функции бинарного поиска: ", ))
def binary_search_iterativ(arr, desired_y):
    # находим среднее значение массива, обозначим верхнюю и нижнюю границу
    middle = len(arr) // 2
    bottom_line = 0
    upper_line = len(arr) - 1
    # сравниваем искомое значение с серединой
    while arr[middle] != desired_y and bottom_line <= upper_line:
        if desired_y > arr[middle]:
            bottom_line = middle + 1
        else:
            upper_line = middle - 1
        # вычесляем индекс новой середины после сдвига одной из границ
        middle = (bottom_line + upper_line) // 2
    # проверка условия, если границы пересеклись
    if bottom_line > upper_line:
        return 'В сгенерированном массиве нет элемента с таким значением'
    else:
        return middle

print(binary_search_iterativ(A, y))
# вставка и удаление в бинарном
a = int(input("Введите элемент, который хотите вставить: ", ))
print("Вывод сортированного массива со вставленным элементом: ")
print(insert_insert(A,a))
b = int(input("Введите элемент, который хотите удалить: ", ))
print("Вывод сортированного массива с удаленным элементом: ")
print(delete_delete(A,b))

# поиск через бинарное дерево 
# Здесь происходит создание корневого узла
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

# Сортировка дерева
def inorder(root):
    if root is not None:
        # Обход левого поддерева
        inorder(root.left)
        # Обход корня
        print(str(root.data) + " ", end=' ')
        # Обход правого поддерева
        inorder(root.right)

# Вставка элемента
def insert(node, data):
    # Если дерево пустое, то вернуть новый узел
    if node is None:
        return Node(data)
    # Переход в левое/правое поддерево и вставка узла
    if data < node.data:
        node.left = insert(node.left, data)
    else:
        node.right = insert(node.right, data)
    return node

# Поиск inorder-преемника
def minValueNode(node):
    current = node
    # Inorder-преемник - это найденный крайний левый лист
    while(current.left is not None):
        current = current.left
    return current

# Удаление узла
def deleteNode(root, data):
    # Если дерево пустое, возвращаем
    if root is None:
        return root
    # Необходимо найти узел, который нужно удалить
    if data < root.data:
        root.left = deleteNode(root.left, data)
    elif(data > root.data):
        root.right = deleteNode(root.right, data)
    else:
        # Если у узла только один дочерний узел или вообще их нет
        if root.left is None:
            temp = root.right
            root = None
            return temp
        elif root.right is None:
            temp = root.left
            root = None
            return temp
        # В случае когда у узла два дочерних узла, помещаем центрированного преемника на место узла, который нужно удалить
        temp = minValueNode(root.right)
        root.data = temp.data
        # Удаляем inorder-преемниа
        root.right = deleteNode(root.right, temp.data)
    return root

root = None
for i in range(len(A)):
    root = insert(root, A[i])

print("\nПоиск через бинарное дерево этого же массива: ", end=' ')
inorder(root)

print("\nВведите элемент, который хотите удалить: ")
del_value = int(input())
root = deleteNode(root, del_value)
print("Вывод сортированный массив с удаленным элементом: ", end=' ')
inorder(root)

print("\nВведите элемент, который хотите вставить: ")
ins_value = int(input())
root = insert(root,ins_value)
print("Вывод сортированный массив со вставленным элементом: ", end=' ')
inorder(root)

# поиск фибоначчи
z = int(input("Введите элемент, который хотите найти с помощью функции поиска фибоначчи: ", ))
def fibonacci_search(arr, desired_z):
    # первые два числа и третье выраженное суммой прошлых
    first_number = 0
    second_number = 1
    summary_number = first_number + second_number
    # наименьшее число в последовательности
    while summary_number < len(arr):
        first_number = second_number
        second_number = summary_number
        summary_number = first_number + second_number
    index = -1
    # пока в массиве есть элементы и значением суммы>1 делаем следующее
    while (summary_number > 1):
        # i ищем для того, чтобы понять на сколько вниз нам сдвигаться
        i = min(index+first_number, len(arr)-1)
        if arr[i] < desired_z:
            # перемещаем  наши значения на два шага вниз и индекс становится индексом элемента
            summary_number = second_number
            second_number = first_number
            first_number = summary_number - second_number
            index = i
        elif arr[i] > desired_z:
            # перемещаем наши значения на один шаг вниз
            summary_number = first_number
            second_number = second_number - first_number
            first_number = summary_number - second_number
        else:
            return i
    # проверяем значение
    if second_number and index < len(arr)-1 and arr[index+1] == desired_z:
        return index+1
    return 'В сгенерированном массиве нет элемента с таким значением'

print(fibonacci_search(A,z))

# вставка и удаление в фибоначчи
c = int(input("Введите элемент, который хотите вставить: ", ))
print("Вывод сортированного массива со вставленным элементом: ")
print(insert_insert(A,c))
d = int(input("Введите элемент, который хотите удалить: ", ))
print("Вывод сортированного массива с удаленным элементом: ")
print(delete_delete(A,d))

# интерполяционный поиск
w = int(input("Введите элемент, который хотите найти с помощью функции интерполяционного поиска: ", ))
def interpolation_searsh(arr, desired_w):
    start_index = 0
    last_index = len(arr)-1
    while start_index <= last_index and desired_w >= arr[start_index] and desired_w <= arr[last_index]:
        # вычисление вероятной позиции искомого элемента
        index = start_index + int(((float(last_index-start_index)/(arr[last_index]-arr[start_index]))*(desired_w-arr[start_index])))
        # элемент найден
        if arr[index] == desired_w:
            return index
        # пересчитываем индекс для праваого подмассива
        elif arr[index] < desired_w:
            start_index = index + 1
        # пересчитываем индекс для левого подмассива
        else:
            last_index = index - 1
    return 'В сгенерированном массиве нет элемента с таким значением'

print(interpolation_searsh(A,w))

# вставка и удаление в интерполяционном поиске
e = int(input("Введите элемент, который хотите вставить: ", ))
print("Вывод сортированного массива со вставленным элементом: ")
print(insert_insert(A,e))
f = int(input("Введите элемент, который хотите удалить: ", ))
print("Вывод сортированного массива с удаленным элементом: ")
print(delete_delete(A,f))
```
